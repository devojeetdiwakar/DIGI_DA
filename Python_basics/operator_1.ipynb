```python
#Arithematic Operator
a, b = 10, 20
c= a+b
d=a-b
e= a*b
f=a/b
g=a//b
h=a%b   
k=a**b
print(c, d, e, f, g, h, k)
```

```python
#Assignment Operator
a = 10
a += 5
print(a)
a -= 5
print(a)
a *= 5
print(a)
a /= 5
print(a)
a //= 5
print(a)
a %= 5
print(a)
a **= 5
print(a)
```

```python
#Comparison Operator
print(a == b)
print(a!=b)
print(a>b)
print(a<b)
print(a>=b)
print(a <= b)
```

```python
#Logical Operator
print(a > 5 and b > 15) #both true only then true it gives as a result
print(a > 5 or b > 15) #at least one condition is true then it gives true as a result
print(not(a > 5 and b > 15)) #not operator gives the opposite of the result of the condition. if the condition is true then it gives false and if the condition is false then it gives true as a result        
True and True
True and False
False and True
False and False
True or True
True or False
False or True
False or False
not True
not False
```
