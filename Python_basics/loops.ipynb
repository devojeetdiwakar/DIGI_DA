```python
n = 123456
rev = 0
while n>0:
    d = n % 10
    rev = rev * 10 + d
    n = n // 10
print(rev)   
```

```python
#prime numnber ---1

n = 13
c = 0
for i in range(1, 14):
    if n % i == 0:
        c+=1

if c == 2:
    print("Number is prime")
else:
    print("Number is not prime")




#prime numnber ---2

a = int(input("Enter a number: "))
for i in range(2, a):
    if a % i == 0:
        print("Number is not prime")
        break
else:    print("Number is prime")
```
