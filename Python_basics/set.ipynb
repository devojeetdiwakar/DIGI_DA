```python
# Set - no duplicates, unordered, mutable values, no indexing, no slicing, no methods like append, insert, pop, remove, clear, sort, reverse, count, index, copy            


# A set is a collection of unique elements that is unordered and mutable. It does not allow duplicate values and does not maintain any specific order of elements. Sets do not support indexing or slicing, and they do not have methods like append, insert, pop, remove, clear, sort, reverse, count, index, or copy that are available for lists.      
# It is unordered, which means that the elements in a set do not have a specific order and cannot be accessed using an index. It is mutable, which means that you can add or remove elements from a set after it has been created. However, since sets do not allow duplicate values, if you try to add an element that already exists in the set, it will not be added again. Sets are commonly used for operations like union, intersection, and difference between collections of data.        


# HOW TO CREATE A SET
A = {23, 45, 8, 4, 4, 8}
print(A) #print the set, duplicates will be removed and the order may not be preserved


my_set = {1, 2, 3, 4, 5} #create a set using curly braces
print(my_set) #print the set

my_set = set([1, 2, 3, 4, 5]) #create a set using the set() constructor with a list as an argument
print(my_set) #print the set

my_set = set() #create an empty set using the set() constructor
print(my_set) #print the empty set

my_set.add(1) #add an element to the set using add() method
print(my_set) #print the set after adding an element
my_set.add(2) #add another element to the set using add() method
print(my_set) #print the set after adding another element

#predefined set functions

A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

print(A.union(B)) #union of two sets A and B, returns a new set that
# contains all the unique elements from both sets A and B

print(A.intersection(B)) #intersection of two sets A and B, returns a new set
# that contains only the elements that are present in both sets A and B

print(A.difference(B)) #difference of two sets A and B, returns a new set
# that contains the elements that are present in set A but not in set B

print(B.difference(A)) #difference of two sets B and A, returns a new set   
# that contains the elements that are present in set B but not in set A 
```

```python
# Difference

A = {10, 20, 30}
B = {30, 40, 50}
C = {50, 60, 70}

print(A.difference(B))
print(A - B)

print(B.difference(A))
print(B - A)   

print(A.difference(C))
print(A - C)

print(A.difference(B).difference(C))
print(A - B - C)

print(C.difference(B).difference(A))
print(C - B - A)
```

```python
# Symmetric Difference

A = {10, 20, 30}
B = {30, 40, 50}
C = {50, 60, 70}

print(A.symmetric_difference(B))
print(A ^ B)        

print(A.symmetric_difference(B).symmetric_difference(C))
print(A ^ B ^ C)

print(A.symmetric_difference(C))
print(A ^ C)

print(B.symmetric_difference(C))
print(B ^ C)        
```

```python
# some functions to try out - update, 

W = {10, 20, 30}
X = {30, 20}
Y = {50, 60, 70}  

#union ka update version - A.update(B) will update set A with the union of sets A and B, meaning it will add all elements from set B to set A that are not already present in set A. If there are any duplicate elements between sets A and B, they will only be added once to set A. After the update operation, set A will contain all unique elements from both sets A and B.
print(W.update({40, 50})) #update set A with elements from another set or iterable
print(W.update(X))
print(W)

#intersection ka update version - A.intersection_update(B) will update set A with the intersection of sets A and B, meaning it will keep only the elements in set A that are also present in set B. After the update operation, set A will contain only the elements that are common to both sets A and B.      
print(W.intersection_update(X))

#difference ka update version - A.difference_update(B) will update set A with the difference of sets A and B, meaning it will remove all elements from set A that are also present in set B. After the update operation, set A will contain only the elements that are present in set A but not in set B.   
print(W.symmetric_difference_update(X))


```
