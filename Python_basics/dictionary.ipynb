```python
alpha = {'a' : 'apple', 'b' : 'ball', 'c' : 'cat', 'd' : 'dog'}

print(alpha['a']) #accessing value of key 'a' in the dictionary alpha
print(alpha['c'])

print(alpha.get('d')) #accessing value of key 'd' in the dictionary alpha using get() method    

alpha['a'] = 'avocado' #updating value of key 'a' in the dictionary alpha
alpha['c']
alpha.get('d')
alpha['a']

alpha['a'] = 'apple'
alpha.get('z', 'nahi mila') #accessing value of key 'c' in the dictionary alpha using get() method with default value 'nahi mila' if key 'c' is not found in the dictionary alpha
```

```python
alpha['b'] = 'ball'
alpha.pop('b') #removing key 'b' and its corresponding value from the dictionary alpha and returning the value of key 'b'
print(alpha)

alpha['b'] = 'ball'
print(alpha)
```

```python
# without using copy function then joh bhi main mai changes honge voh a par bhi reflect hoga kyuki it is linking wioth the original
a = alpha
print(a)

# with using copy function then joh bhi main mai changes honge voh a par reflect nahi hoga kyuki it is not linking wioth the original   
b = alpha.copy()
print(b)
```

```python
alpha['z'] = 'zebra' #adding new key-value pair 'z' : 'zebra' to the dictionary alpha
alpha['z'] #accessing value of key 'z' in the dictionary alpha
```

```python
alpha.items() #returns a view object that displays a list of dictionary's key-value tuple pairs

for pair in alpha.items(): #iterating through the dictionary alpha using items() method to access both keys and values as tuples
    print(pair)   


alpha.keys() #returns a view object that displays a list of dictionary's keys

for key, value in alpha.items(): #iterating through the dictionary alpha using items() method to access both keys and values
    print(f"Key: {key}, Value: {value}")             


alpha.values() #returns a view object that displays a list of dictionary's values

for value in alpha.values(): #iterating through the dictionary alpha using values() method to access only values
    print(f"Value: {value}")
  
```

```python
smartphone = {
        'brand': 'Apple', 
        'model': 'iPhone 13', 
        'colors': ['black', 'white', 'blue', 'cosmic orange'], 
        'storage': '128GB', 
        'price': 170000
    }
print(smartphone)
print(smartphone['colors'])
print(smartphone['colors'][3]) #accessing the 4th color option of the smartphone using nested indexing 
print(smartphone['colors'][-1])
print(smartphone['brand'])

print(smartphone.get('model')) #accessing value of key 'model' in the dictionary smartphone using get() method
print(smartphone.get('battery', 'Battery information not available')) #accessing value of key 'battery' in the dictionary smartphone using get() method with default value 'Battery information not available' if key 'battery' is not found in the dictionary smartphone           
print(smartphone.get(smartphone['brand'])) #accessing value of key 'Apple' in the dictionary smartphone using get() method with nested key access   
print(smartphone.get('colors')) 


print(smartphone.get('colors')[3])
print(smartphone.get('colors')[3][4])  
print(smartphone.get('colors')[3].split()) #accessing the first word of the 3rd color option of the smartphone using nested indexing and split() method
print(smartphone.get('colors')[3].split()[0])
print(smartphone.get('colors')[3].split()[1])

```

```python
smartphone['colors'].append('orange') #adding 'orange' to the list of colors in the dictionary smartphone
print(smartphone['colors'])

smartphone['colors'].remove('orange') #removing 'cosmic orange' from the list of colors in the dictionary smartphone
print(smartphone['colors'])
```
