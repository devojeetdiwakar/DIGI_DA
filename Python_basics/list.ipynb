```python
movie = ['Thor', 'Batman', 'X-men', 'Ironman', 'Krish', 'Dhurandar']
```

```python
#indexing 
print(movie[0]) #first element of the list
print(movie[1]) #second element of the list
print(movie[-1]) #last element of the list
print(movie[-2]) #second last element of the list
print(movie[0:3]) #first 3 elements of the list
print(movie[1:4]) #elements from index 1 to index 3
print(movie[::2]) #every second element of the list
print(movie[1::2]) #every second element of the list starting from index 1
print(movie[::-1]) #reverse the list
```

```python
#append (last mai add karne keliye) , inset(inbetween mai add karne ke liya), extend
print(movie.append('Avengers')) #append a new element to the end of the list
print(movie) #print the list after appending a new element  
print(movie.insert(2, 'Spiderman')) #insert a new element at index 2
print(movie) #print the list after inserting a new element at index 2   
print(movie.extend(['Captain America', 'Hulk'])) #extend the list by adding multiple elements to the end of the list
print(movie) #print the list after extending it by adding multiple elements to the end of the list
```

```python
#pop (last element ko remove karne ke liye), clear (sari list ko empty karne ke liye)
print(movie.pop()) #remove the last element of the list and return it
print(movie) #print the list after popping the last element
print(movie.pop(2)) #remove the element at index 2 and return it
print(movie) #print the list after popping the element at index 2
print(movie) #print the list after removing the element 'Krish'
print(movie.clear()) #clear the list and make it empty
print(movie) #print the list after clearing it  
```

```python
movie = ['Thor', 'Batman', 'X-men', 'Ironman', 'Krish', 'Dhurandar']
print(movie.count('Ironman')) #count the number of occurrences of 'Ironman' in the list
movie.reverse() #reverse the list
print(movie) #print the list after reversing it
print(movie.index('Ironman'))
```

```python
#SORTING ----  1

movie = ['Thor', 'Batman', 'X-men', 'Ironman', 'Krish', 'Dhurandar']
print(movie.index('Ironman'))
print(movie.sort(reverse=True)) #sort the list in reverse order (descending order) and return None
print(movie)
print(movie.sort(reverse=False)) #sort the list in ascending order and return None
print(movie)
print(movie.sort()) 
print(movie)
      
```

```python
#SORTING ----  2

movie = ['Thor', 'Batman', 'X-men', 'Ironman', 'Krish', 'Dhurandar']
print(sorted(movie)) #sort the list in ascending order and return a new sorted list without modifying the original list
print(movie) #print the original list after sorting it using sorted() function  
print(sorted(movie, reverse=False)) #sort the list in ascending order and return a new sorted list without modifying the original list
print(movie) #print the original list after sorting it in ascending order using sorted() function   
sorted(movie, reverse=True) #sort the list in reverse order (descending order) and return a new sorted list without modifying the original list
print(movie) #print the original list after sorting it in reverse order using sorted() function   
```

```python
# COPYYY

nums = [1, 2, 3, 4, 5]

nums_copy = nums #create a copy of the list by assigning it to a new variable
print('Original list:', nums)
print('Copied list:', nums_copy) 
nums.append(6) #append a new element to the copied list
print('Original list:', nums) #print the original list to show that it is modified after
print('Copied list:', nums_copy) #print the copied list after appending a new element



```
