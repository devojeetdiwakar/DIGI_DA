```python
print(movie[0:20:2])
print(movie[:])
print(movie[::2])
print(movie[::-1]) # reverse the string
print(name2[0:10:-1])
print(name2[10:0:-1]) # reverse the string from index 10 to 1 and always write the start index greater than the end index when using negative step
#in reverse slicing, the start index is always greater than the end index and the step is negative. If the step is positive, then the start index should be less than the end index.
#[::-1] -1 is used for reverse

print(name2[4]) #SPACE if we write name2[4], then it will print the character at index 4, which is 'S'. But if we write name2[4:5], then it will print the character at index 4 and the end index is not included in the output. So, it will print only the character at index 4, which is 'S'.
print(name2[4:5]) #SPACE if need to print a single character, then the end index should be one more than the start index. If we write name2[4], then it will print the character at index 4, but if we write name2[4:5], then it will print the character at index 4 and the end index is not included in the output. So, it will print only the character at index 4.
```

```python
print(name2.lower()) #lowercase
print(name2.upper())#uppercase
print(name2.title()) #title case
print(name2.capitalize()) #capitalize the first character of the string and the rest of the characters will be in lowercase
print(name2.swapcase()) #swapcase the string, it will convert uppercase to
```

```python
#Alignment functions
print(name2.rjust(30)) #right justify the string
print(name2.ljust(30)) #left justify the string
print(name2.center(30)) #center the string
print(len(name2.rjust(30)))
print(len(name2.ljust(30)))
print(len(name2.center(30)))
print(name2.rjust(30, "*")) #right justify the string with *
print(name2.ljust(30, "*")) #left justify the string with *
print(name2.ljust(30, "a")) #left justify the string with a, it will fill the remaining space with a
print(name2.center(30, "-")) #center the string with *
print((' '+name2+' ').center(30, "-"))
```

```python
print(name2.isupper()) #check if the string is in uppercase or not, it will return False because the string is not in uppercase
print(name2.islower()) #check if the string is in lowercase or not, it will return False because the string is not in lowercase
print(name2.isalpha()) #check if the string contains only alphabets or not, it will return False because the string contains space
print(name2.isalnum()) #check if the string contains only alphanumeric characters or not, it will return False because the string contains space
print(name2.isnumeric()) #check if the string contains only numeric characters or not, it will return False because the string contains alphabets and space    
print(name2.isdigit()) #check if the string contains only digits or not, it will return False because the string contains alphabets and space
print(name2.istitle()) #check if the string is in title case or not, it will return True because the string is in title case
print(name2.isspace()) #check if the string contains only space or not, it will return False because the string contains alphabets and space


print('123.3'.isnumeric())
print('abc 123.3'.isalnum())
```

```python
print(name2.count('e')) #count the number of occurrences of a character in the string, it will return 2 because there are 2 'a' in the string
print(name2.index('n',4)) #find the index of the first occurrence of a character in the string starting from index 4, it will return 10 because 'n' is at index 10 and the search starts from index 4
print(name2.index('n')) # find the index of the first occurrence of a character in the string, it will return 1 because 'n' is at index 1 
print(name2.rindex('n')) #find the index of the last occurrence of a character in the string, it will return 10 because 'n' is at index 10 and it is the last occurrence of 'n' in the string
print(name2.find('n',4)) #find the index of the first occurrence of a character in the string starting from index 4, it will return 10 because 'n' is at index 10 and the search starts from index 4
print(name2.find('n')) # find the index of the first occurrence of a character in the string, it will return 1 because 'n' is at index 1
print(name2.rfind('n')) #find the index of the last occurrence of a character in the string, it will return 10 because 'n' is at index 10 and it is the last occurrence of 'n' in the string
print(name2.find('a')) #find the index of the first occurrence of a character in the string, it will return 5 because 'a' is at index 5 and it is the first occurrence of 'a' in the string
```

```python
print(name2.replace('n', 'N')) #replace a character in the string with another character, it will replace all the occurrences of 'n' with 'N' in the string
print(name2.replace('n', 'N', 1)) #replace a character in the string with another character, it will replace only the first occurrence of 'n' with 'N'
print(name2.replace('n', 'N', 2)) #replace a character in the string with another character, it will replace only the first 2 occurrences of 'n' with 'N'       
```

```python
print(name2.split()) #split the string into a list of words, it will split the string by space and return a list of words
print(name2.split(' ')) #split the string by space and return a list of words, it will split the string by space and return a list of words 
print('https://meet.google.com/yun-rdf-dfs'.split('/')) #split the string by '/' and return a list of words, it will split the string by '/' and return a list of words         
print(name2.split('o')) #split the string by 'o' and return a list of words
print(name2.split('o', 1)) #split the string by 'o' and return a list of words, it will split the string only once and return a list of 2 words
print(name2.split('o', 2)) #split the string by 'o' and return a list of words, it will split the string only twice and return a list of 3 words    
```
